```python
import cv2
import os
#import datetime
import time
from pathlib import Path
import matplotlib.pyplot as plt

# plate_cascade = cv2.CascadeClassifier(r'HAARcascadeImagensStreetView26615P35658N 02.09.2021.xml') #USADO NO PRIMEIRO E SEGUNDO ARTIGO
plate_cascade = cv2.CascadeClassifier(r'cascade - 17962 positivas 35658 negativas - ARTIGO 3/cascade.xml')
# plate_cascade = cv2.CascadeClassifier(r'cascade lapisco/cascade.xml')

def adjusted_detect_plate(img):
     
    plate_img = img.copy()
#    start = datetime.datetime.now() #time.clock()
#     plate_rect = plate_cascade.detectMultiScale(plate_img, scaleFactor=1.1, minNeighbors=1, minSize=(1,1), maxSize=(500,500)) #USADO NO PRIMEIRO E SEGUNDO ARTIGO
    plate_rect = plate_cascade.detectMultiScale(plate_img, scaleFactor=1.06, minNeighbors=1, minSize=(1,1), maxSize=(500,500)) #USADO NO TERCEIRO ARTIGO
#    plate_rect = plate_cascade.detectMultiScale(plate_img, scaleFactor=1.045, minNeighbors=2, minSize=(40,13), maxSize=(180,58)) #USAR NO QUARTO ARTIGO
#    end = datetime.datetime.now()
#     result = end - start
#     print(result)
#     soma += result
    cropped = False
    crop = None
    for (x, y, w, h) in plate_rect:
        cv2.rectangle(plate_img, (x, y), (x + w, y + h), (0, 0, 255), 3)
        crop = img[y:y+h, x:x+w]
        cropped = True  
    return plate_img, cropped, crop

def read():
#     folder = 'Ajuste de Perspectiva\Table6datasetMercosulReal1000Lote1'
#     pathToSave = r'Ajuste de Perspectiva\Artigo 3\Resutados\Table6\result1000'
#     folder = r'Ajuste de Perspectiva\cropsAdjusted'
    folder = r'crops'
    pathToSave = r'Ajuste de Perspectiva\cropsAdjustedDetectionTimeCalculation'
    
#     pathToSave = 'resultCascadeSpecialAdjusted'
#     folder = 'cropsAdjusted'
    images = []
    i = 0
    tempoGasto = 0
    for filename in os.listdir(folder):
#         print(filename.split('.')[0])
        img = cv2.imread(os.path.join(folder,filename))
#         plt.imshow(img), plt.xticks([]), plt.yticks([])
#         plt.show()
        if img is not None:
            i +=1
            start = time.time()
            imgToSave, cropped, crop = adjusted_detect_plate(img)
            if cropped:
#                 cv2.imwrite(os.path.join(pathToSave , str(i)+'.jpg'), imgToSave)
                crop = cv2.resize(crop,(240,78))
#                 crop = cv2.resize(crop,(153,50))
#                 cv2.imwrite(os.path.join(pathToSave , str(i)+'.jpg'), crop)
                cv2.imwrite(os.path.join(pathToSave , filename.split('.')[0]+'.jpg'), crop)
#                 cv2.waitKey(0)
                end = time.time()
                result = end - start
                print(result)
                tempoGasto += result
    print('total: ',  i)
    print('Tempo total: ',  tempoGasto)
    print('Tempo Gasto Por Detecção: ',  tempoGasto / 100)

read()                
```
