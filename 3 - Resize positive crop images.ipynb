# Crop a imagem a partir dos valores do txt

```python
import cv2
import os
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt


labelFolder = 'labelReal'
imageFolder = 'datasetReal'
pathToSave = 'p'
i = 0
for filename in os.listdir(labelFolder):

    arquivo = open(labelFolder+'/'+filename, 'r')
    
    for linha in arquivo:    
        
        imageName = filename.replace('.txt', '.png')
        image = cv2.imread(os.path.join(imageFolder,imageName))

        if image is not None:
            dh, dw, _ = image.shape
            _, x, y, w, h = map(float, linha.split(' '))

            l = int((x - w / 2) * dw)
            r = int((x + w / 2) * dw)
            t = int((y - h / 2) * dh)
            b = int((y + h / 2) * dh)

            if l < 0:
                l = 0
            if r > dw - 1:
                r = dw - 1
            if t < 0:
                t = 0
            if b > dh - 1:
                b = dh - 1
            
            # Aumentar o crop da imagem quando poss√≠vel
#             if t - 5 < 0:
#                 t = 0 
#             else:
#                 t = t - 5
            
#             if l - 5 < 0:
#                 l = 0 
#             else:
#                 l = l - 5
#             b = b + 5      
#             r = r + 5
        
            crop = image[t:b,l:r]
            crop = cv2.cvtColor(crop, cv2.COLOR_BGR2GRAY)
            
#             ORIGINAIS DOS ARTIFOS ANTERIORES AO TERCEIRO
#             imageName = '40x13' + imageName
#             crop = cv2.resize(crop,(40,13))
#             imageName = '80x29' + imageName
#             crop = cv2.resize(crop,(80,29))
#             imageName = '120x39' + imageName
#             crop = cv2.resize(crop,(120,39))
#             imageName = '180x59' + imageName
#             crop = cv2.resize(crop,(180,59))
#             imageName = '240x78' + imageName
#             crop = cv2.resize(crop,(240,78))
#             imageName = '360x107' + imageName
#             crop = cv2.resize(crop,(360,107))
#             imageName = '420x137' + imageName
#             crop = cv2.resize(crop,(420,137))

#             TERCEIRO ARTIGO
#             imageName = '60x19' + imageName
#             crop = cv2.resize(crop,(60,19))
#             imageName = '80x26' + imageName
#             crop = cv2.resize(crop,(80,26))
#             imageName = '100x32' + imageName
#             crop = cv2.resize(crop,(100,32))
#             imageName = '120x39' + imageName
#             crop = cv2.resize(crop,(120,39))
#             imageName = '140x45' + imageName
#             crop = cv2.resize(crop,(140,45))
            imageName = '160x52' + imageName
            crop = cv2.resize(crop,(160,52))


            
#             cv2.imwrite(os.path.join(pathToSave , imageName), crop)
            cv2.imwrite(os.path.join(pathToSave , imageName.replace(" ", "_").replace("(","").replace(")","").split(".")[0]+".jpg"), crop)
            cv2.waitKey(0)
#             cv2.rectangle(image, (l, t), (r, b), (255, 0, 0), 5)

#             plt.imshow(crop)
#             plt.show()
print('FIM')
```
